```python
#@markdown <center><h2><font color=yellow>
import requests
import os
from tqdm import tqdm
import urllib.parse
import re
import logging  # Import the logging module
# --- Logging Setup ---
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def write_failed_downloads_to_file(failed_downloads, downloader_name, download_directory):
    """Writes the list of failed URLs and their corresponding filenames (or extracted names) to a text file."""
    if not failed_downloads:
        return  # Nothing to write

    file_path = os.path.join(download_directory, f"failed_downloads_{downloader_name}.txt")
    try:
        with open(file_path, "w") as f:
            for url in failed_downloads:
                # Try to extract a filename; use a placeholder if it fails
                file_name = extract_filename_from_url(url) or "unknown_filename"
                f.write(f"{url}, {file_name}\n")
        logging.info(f"List of failed {downloader_name} downloads saved to: {file_path}")
    except Exception as e:
        logging.error(f"Error writing failed downloads to file: {e}")


# --- nzbCloud (requests with URL and Filename input) ---

def download_files_nzbcloud(urls, filenames, cf_clearance, download_directory):
    """Downloads files using requests, tqdm progress bar, and cf_clearance cookie (nzbCloud)."""
    cookies = {"cf_clearance": cf_clearance}
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Referer": "https://app.nzbcloud.com/",  # Adjust as needed
    }
    url_filename_pairs = list(zip(urls, filenames))
    failed_downloads = []  # Keep track of failed URLs

    for idx, (url, file_name) in enumerate(url_filename_pairs):
        url = url.strip()
        file_name = file_name.strip()
        if not url or not file_name:
            logging.warning(f"Skipping (nzbCloud): URL or filename is missing for pair {idx + 1}.")
            continue

        try:
            response = requests.get(url, headers=headers, cookies=cookies, stream=True, timeout=30)
            response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
            total_size = int(response.headers.get('content-length', 0))
            block_size = 1024

            logging.info(f"Downloading (nzbCloud) {file_name}...")
            with tqdm(total=total_size, unit='B', unit_scale=True, desc=f"Downloading {file_name}", position=0,
                      leave=True) as t:
                full_file_path = os.path.join(download_directory, file_name)
                os.makedirs(os.path.dirname(full_file_path), exist_ok=True)  # Ensure directory exists
                with open(full_file_path, "wb") as file:
                    for chunk in response.iter_content(chunk_size=block_size):
                        if chunk:  # filter out keep-alive new chunks
                            file.write(chunk)
                            t.update(len(chunk))
            logging.info(f"Download complete (nzbCloud): {file_name}")

        except requests.exceptions.RequestException as e:
            logging.error(f"Failed to download (nzbCloud) {file_name}: {e}")
            failed_downloads.append(url)  # Add the failed URL to the list
            if isinstance(e, requests.exceptions.Timeout):
                logging.info(" (The request timed out. أ¢عˆآ³)")
            elif isinstance(e, requests.exceptions.HTTPError):
                logging.info(f" (HTTP Error: {response.status_code} - {response.reason} ظ‹ع؛ع‘آ«)")

        except Exception as e:
            logging.exception(f"An unexpected error occurred (nzbCloud) while downloading {file_name}: {e}")
            failed_downloads.append(url)  # Add the failed URL to the list

    write_failed_downloads_to_file(failed_downloads, "nzbcloud", download_directory)


# --- DeltaLeech (filename extraction and cleaning) ---

def clean_filename(filename):
    """Cleans filenames by replacing invalid characters with dots and collapsing multiple dots."""
    if not isinstance(filename, str):
        logging.error(f"Invalid input to clean_filename: expected string, got {type(filename)}. Using default name.")
        return "downloaded_file_invalid_input"

    cleaned_filename = filename # Start with the original name

    try:
        # Decode URL-encoded characters like %20 -> space
        cleaned_filename = urllib.parse.unquote(cleaned_filename, encoding='utf-8')
    except Exception as e:
        logging.warning(f"Error during URL decoding of filename '{filename}': {e}. Proceeding with potentially encoded name.")
        # Proceed with the potentially still encoded filename

    # --- STEP 1 ---
    # Replace any character that is NOT a letter, number, underscore, or dot with a SINGLE dot.
    try:
        cleaned_filename = re.sub(r'[^\w\.]', '.', cleaned_filename)
    except Exception as e:
        logging.error(f"Error during regex substitution 1 on '{cleaned_filename}': {e}")

    # --- STEP 2 ---
    # Replace sequences of 2 or more dots with a single dot. (THIS SHOULD BE ENOUGH)
    try:
        cleaned_filename = re.sub(r'\.{2,}', '.', cleaned_filename)
    except Exception as e:
        logging.error(f"Error during regex substitution 2 on '{cleaned_filename}': {e}")

    # --- STEP 3 ---
    # Remove leading or trailing dots or underscores
    cleaned_filename = cleaned_filename.strip('._')

    # --- STEP 4 (Optional Underscore Removal) ---
    # Remove this line if you want to KEEP underscores.
    cleaned_filename = cleaned_filename.replace("_", ".")

    # --- FINAL CHECK FOR EMPTY FILENAME ---
    if not cleaned_filename:
        logging.warning(f"Filename became empty after cleaning original: '{filename}'. Using default name.")
        cleaned_filename = "downloaded_file"

    # --- STEP 5: REDUNDANT FINAL DOT COLLAPSE (Added as requested) ---
    # This step is added because problems persist, but Step 2 should have already done this.
    # If double dots still exist before this line, the issue is likely that the code isn't updated correctly.
    try:
        logging.debug(f"Filename before final redundant collapse: '{cleaned_filename}'") # Optional debug log
        cleaned_filename = re.sub(r'\.{2,}', '.', cleaned_filename)
        logging.debug(f"Filename after final redundant collapse: '{cleaned_filename}'") # Optional debug log
    except Exception as e:
        logging.error(f"Error during final redundant regex substitution on '{cleaned_filename}': {e}")
    # --- END OF ADDED STEP ---

    return cleaned_filename
# --- END OF FUNCTION TO PASTE ---


def extract_filename_from_url(url):
    """Extracts and cleans the filename from a given URL."""
    try:
        parsed_url = urllib.parse.urlparse(url)
        path = parsed_url.path
        filename = os.path.basename(path)
        return clean_filename(filename)
    except Exception as e:
        logging.warning(f"Error extracting filename from URL: {e}")
        return None


def download_file_deltaleech(url, file_name, cf_clearance, download_directory):
    """Downloads a single file with improved error handling and progress bar."""
    cookies = {"cf_clearance": cf_clearance} if cf_clearance else {}
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Referer": url
    }

    try:
        response = requests.get(url, headers=headers, cookies=cookies, stream=True, timeout=30)
        response.raise_for_status()

        total_size = int(response.headers.get('content-length', 0))
        block_size = 1024

        full_file_path = os.path.join(download_directory, file_name)
        os.makedirs(os.path.dirname(full_file_path), exist_ok=True)

        with open(full_file_path, "wb") as file, tqdm(
                total=total_size,
                unit='B',
                unit_scale=True,
                desc=f"Downloading ظ‹ع؛ع‘â‚¬ {os.path.basename(file_name)}",
        ) as progress_bar:
            for chunk in response.iter_content(chunk_size=block_size):
                if chunk:
                    file.write(chunk)
                    progress_bar.update(len(chunk))
            logging.info(f"{file_name} Download complete! ظ‹ع؛عکâ€°")

    except requests.exceptions.RequestException as e:
        logging.error(f"Failed to download {file_name}: {e}")
        if isinstance(e, requests.exceptions.Timeout):
            logging.info(" (The request timed out. أ¢عˆآ³)")
        elif isinstance(e, requests.exceptions.HTTPError):
            logging.info(f" (HTTP Error: {response.status_code} - {response.reason} ظ‹ع؛ع‘آ«)")
        return False  # Indicate failure
    except Exception as e:
        logging.exception(f"An unexpected error occurred while downloading {file_name}: {e}")
        return False  # Indicate failure

    return True  # Indicate success


def download_multiple_files_deltaleech(urls, file_names, cf_clearance, download_directory):
    """Downloads multiple files from DeltaLeech, handling potential errors."""
    if len(urls) != len(file_names):
        logging.error("Error: The number of URLs and filenames must match! ظ‹ع؛â€œعˆ")
        return

    failed_downloads = []
    for url, file_name in zip(urls, file_names):
        if not download_file_deltaleech(url, file_name, cf_clearance, download_directory):
            failed_downloads.append(url)

    write_failed_downloads_to_file(failed_downloads, "deltaleech", download_directory)



# --- Bitso Downloader ---
def download_file_bitso(url, file_name, download_directory, referer_url="", _identity_value="", phpsessid_value=""):
    """Downloads a file from bitso.ir, handling cookies and referer."""
    cookies = {}
    if _identity_value:
        cookies["_identity"] = urllib.parse.quote_plus(_identity_value)
    if phpsessid_value:
        cookies["PHPSESSID"] = urllib.parse.quote_plus(phpsessid_value)

    headers = {
        "Referer": referer_url,
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.9999.99 Safari/537.36"
    }

    try:
        response = requests.get(url, headers=headers, cookies=cookies, stream=True, timeout=60)
        response.raise_for_status()

        total_size = int(response.headers.get('content-length', 0))
        block_size = 1024
        full_file_path = os.path.join(download_directory, file_name)
        os.makedirs(os.path.dirname(full_file_path), exist_ok=True)

        with open(full_file_path, "wb") as file, tqdm(
                total=total_size, unit='B', unit_scale=True, desc=f"Downloading ظ‹ع؛ع‘â‚¬ {file_name}"
        ) as progress_bar:
            for chunk in response.iter_content(chunk_size=block_size):
                if chunk:
                    file.write(chunk)
                    progress_bar.update(len(chunk))
            logging.info(f"Download complete: {file_name} ظ‹ع؛عکâ€°")

    except requests.exceptions.RequestException as e:
        logging.error(f"Failed to download {file_name}: {e}")
        if isinstance(e, requests.exceptions.Timeout):
            logging.info(" (The request timed out. أ¢عˆآ³)")
        elif isinstance(e, requests.exceptions.HTTPError):
            logging.info(f" (HTTP Error: {response.status_code} - {response.reason} ظ‹ع؛ع‘آ«)")
        return False
    except Exception as e:
        logging.exception(f"An unexpected error occurred while downloading {file_name}: {e}")
        return False

    return True


def download_multiple_files_bitso(urls, file_names, download_directory, referer_url="", _identity_value="", phpsessid_value=""):
    """Downloads multiple files from bitso.ir."""
    if len(urls) != len(file_names):
        logging.error("Error: The number of URLs and filenames must match! ظ‹ع؛â€œعˆ")
        return

    failed_downloads = []
    for url, file_name in zip(urls, file_names):
        if not download_file_bitso(url, file_name, download_directory, referer_url, _identity_value, phpsessid_value):
            failed_downloads.append(url)

    write_failed_downloads_to_file(failed_downloads, "bitso", download_directory)


# --- Main Control ---

def main():
    downloader_choice = input("Choose downloader (1: nzbCloud, 2: DeltaLeech, 3: Bitso): ")

    if downloader_choice == "1":
        urls_input = input("Enter URLs, one per line (separated by commas): ")
        urls = [url.strip() for url in urls_input.split(',')]

        cf_clearance = input("Enter Cloudflare clearance cookie (optional): ").strip() or None
        download_directory = input("Enter download directory (default: /content/Upload): ").strip() or "/content/Upload"
        filenames_input = input("Enter filenames, one per line (separated by commas) (nzbCloud): ")
        filenames = [filename.strip() for filename in filenames_input.split(',')]

        if len(urls) != len(filenames):
            logging.error("Error: The number of URLs and filenames must match! ظ‹ع؛â€œعˆ")
            return

        download_files_nzbcloud(urls, filenames, cf_clearance, download_directory)

    elif downloader_choice == "2":
        # This part remains the same - getting the URLs
        urls_input = input("Enter URLs, one per line (separated by commas): ")
        urls = [url.strip() for url in urls_input.split(',')]

        # ---vvv--- MODIFICATION START ---vvv---

        # Comment out or delete the original input prompts:
        # cf_clearance = input("Enter Cloudflare clearance cookie (optional): ").strip() or None
        # download_directory = input("Enter download directory (default: /content/Upload): ").strip() or "/content/Upload"

        # Assign fixed values instead:
        cf_clearance = None  # Set to None (since it was optional) or your specific cookie string if always needed
        download_directory = "/content/hardsub" # @param {type:"string"}
        # @markdown > <i>`in Case of Deltaleech mode` Select your Directory

        # Optional: Add logging to confirm the values being used
        logging.info(f"Using hardcoded cf_clearance for DeltaLeech: {cf_clearance}")
        logging.info(f"Using hardcoded download directory for DeltaLeech: {download_directory}")

        # ---^^^--- MODIFICATION END ---^^^---

        prompt_message = "If you don't wanna change the file names put y (or Enter) " # Changed prompt to show default
        # Get input, strip whitespace, default to 'y' if empty, THEN convert to lowercase
        use_url_filenames_input = (input(prompt_message).strip() or 'y').lower()
        # This line remains the same - it sets the boolean based on whether the result is 'y'
        use_url_filenames = use_url_filenames_input == 'y'

        if use_url_filenames:
            # Using the version that calls clean_filename via extract_filename_from_url
            filenames = [extract_filename_from_url(url) for url in urls]
            if None in filenames:
                logging.error("Some URLs could not have their filename extracted. ظ‹ع؛â„¢ظ¾")
                return
        else:
            filenames_input = input("Enter filenames, one per line (separated by commas) (DeltaLeech): ")
            # Using the version that cleans manually entered filenames
            filenames = [clean_filename(filename.strip()) for filename in filenames_input.split(',') if filename.strip()]

            if len(urls) != len(filenames):
                logging.error("Error: The number of URLs and cleaned filenames must match! ظ‹ع؛â€œعˆ")
                return

        # This part remains the same - calling the download function
        download_multiple_files_deltaleech(urls, filenames, cf_clearance, download_directory)

    elif downloader_choice == "3":
        cf_clearance = None
        download_directory = "/content/uploaded" # @param {type:"string"}
         # @markdown > <i>`in Case of Bitso mode` Select your Directory
        referer_url = "https://panel.bitso.ir/"
        _identity_value = ""
        phpsessid_value = ""

        urls_input = input("Enter URLs, one per line (separated by commas): ")
        urls = []
        for line in urls_input.splitlines():
            line = line.strip()
            if line:
                urls.extend([url.strip() for url in line.split(',') if url.strip()])

        filenames_input = input("Enter filenames manually, one per line (separated by commas) (Bitso): ")
        filenames = []
        for line in filenames_input.splitlines():
            line = line.strip()
            if line:
                filenames.extend([fn.strip() for fn in line.split(',') if fn.strip()])

        for url in urls:
            try:
                result = urllib.parse.urlparse(url)
                if not all([result.scheme, result.netloc]):
                    raise ValueError
            except ValueError:
                logging.error(f"Invalid URL: {url} ظ‹ع؛ع‘آ«")
                return

        if len(urls) != len(filenames):
            logging.error("Error: The number of URLs and filenames must match! ظ‹ع؛â€œعˆ")
            logging.info(f" Number of URLs: {len(urls)} ظ‹ع؛â€‌آ¢")
            logging.info(f" Number of filenames: {len(filenames)} ظ‹ع؛â€‌آ¢")
            return

        download_multiple_files_bitso(urls, filenames, download_directory, referer_url, _identity_value, phpsessid_value)

    else:
        logging.error("Invalid downloader choice. ظ‹ع؛ع‘آ«")


if __name__ == "__main__":
    main()
```

# Video Spliter
<left><img src='https://simp6.jpg5.su/images3/image-Photoroom-37b51f0333175026b.png' height="160" alt="JDownloader-logo"/></left>

```python
# @title <font color=orange> Split Video
# @title Main Code
# @markdown <br><center><h2><font color=white><strong>Split Video Files</strong></center>
import subprocess
import json
import os  # Make sure os is imported
from datetime import datetime
from asyncio import sleep
from tqdm.notebook import tqdm  # Importing tqdm for progress bar in Jupyter/Colab

def get_total_size(directory):
    """Returns the total size of all files within a directory in bytes."""
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(directory):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            # skip if it is symbolic link
            if not os.path.islink(fp):
                total_size += os.path.getsize(fp)
    return total_size

def getSize(file_path):
    """Returns the size of the file in bytes."""
    return os.stat(file_path).st_size

def speedETA(start_time, bytes_processed, total_size):
    """Returns speed, ETA, and percentage completion."""
    elapsed_time = (datetime.now() - start_time).total_seconds()
    speed = bytes_processed / elapsed_time if elapsed_time > 0 else 0
    eta = (total_size - bytes_processed) / speed if speed > 0 else 0
    percentage = (bytes_processed / total_size) * 100 if total_size > 0 else 0
    return speed, eta, percentage

async def splitVideo(input_path, output_dir, max_size, remove: bool):
    """
    Splits the video into multiple parts based on the specified maximum size.
    """

    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Ensure input path is valid
    if not os.path.exists(input_path):
        raise ValueError(f"Error: The file at {input_path} does not exist.")

    _, filename = os.path.split(input_path)
    just_name, extension = os.path.splitext(filename)

    # FFmpeg command to get video information in JSON format
    cmd = ["ffprobe", "-v", "quiet", "-print_format", "json", "-show_format", input_path]

    bitrate = None
    try:
        # Run the command and get output
        output = subprocess.check_output(cmd)
        video_info = json.loads(output)
        # Get bitrate in bits per second
        bitrate = float(video_info["format"]["bit_rate"])
    except subprocess.CalledProcessError:
        print("Error: Could not get video bitrate")
        bitrate = 1000  # Default bitrate if we can't retrieve it

    # Convert target size from MB to bits
    target_size_bits = max_size * 8 * 1024 * 1024

    # Calculate duration in seconds
    duration = int(target_size_bits / bitrate)
    if duration < 1:
        duration = 1  # Ensure at least 1-second segments

    # Output path for split video parts
    output_pattern = os.path.join(output_dir, f"{just_name}.part%03d{extension}")

    # FFmpeg command to split the video
    cmd = f'ffmpeg -i "{input_path}" -c copy -f segment -segment_time {duration} -reset_timestamps 1 "{output_pattern}"'

    print(f"Splitting video: {filename}")
    start_time = datetime.now()

    proc = subprocess.Popen(cmd, shell=True)
    total_size = getSize(input_path)
    total_in_unit = total_size / (1024 * 1024 * 1024)  # Convert to GB

    # Initializing tqdm for progress bar
    progress_bar = tqdm(total=100, desc="Splitting Video", unit="%")

    while proc.poll() is None:
        speed, eta, percentage = speedETA(
            start_time, get_total_size(output_dir), total_size  # Use get_total_size
        )

        # Update progress bar with percentage
        progress_bar.n = percentage
        progress_bar.last_print_n = percentage
        progress_bar.set_postfix({"Speed": f"{speed/1024/1024:.2f} MB/s", "ETA": f"{eta:.0f}s"}) #Added speed and ETA info
        progress_bar.update(0)

        await sleep(1)

    returncode = proc.returncode
    if returncode != 0:
        print(f"FFmpeg process finished with error code: {returncode}") #Error Handling
    progress_bar.n = 100
    progress_bar.last_print_n = 100
    progress_bar.update(0) #Ensure progress bar is 100

    if remove:
        os.remove(input_path)
        print(f"Original file '{filename}' removed.") #Confirmation message

    print(f"Video splitting completed. Output saved in: {output_dir}") #Completion message
    return output_dir  # Returns the directory where the split parts were saved



# User Input for dynamic configuration
input_video = input("Enter the full path of the input video: ")
output_directory = input("Enter the output directory (e.g., /content/Downloads/output): ")
max_size = int(input("Enter the maximum size per split part in MB (e.g., 2000): "))
remove_original = input("Do you want to delete the original video after splitting? (yes/no): ").strip().lower() == "yes"

# Run the function (IMPORTANT: Use in an async environment like Jupyter/Colab)
# In Colab, you can run this directly. In a standard Python script, you'd need:
# import asyncio
# asyncio.run(splitVideo(input_video, output_directory, max_size, remove_original))
await splitVideo(input_video, output_directory, max_size, remove_original)
```

```python
#@markdown # ظ‹ع؛â€œآ¦ Unzip Utility for Google Colab
#@markdown ---
#@markdown ### ظ‹ع؛â€œظ¾ Input & Output Paths
#@markdown Enter the full path to the zip file and the desired extraction directory.
zip_path_param = "/content/eva/Eva.Elfie.OF.Part2.zip" #@param {type:"string"}
extract_to_path_param = "/content/eva/" #@param {type:"string"}
password_param = "" #@param {type:"string"}
#@markdown ---
#@markdown ### أ¢ع‘â„¢أ¯آ¸عˆ Unzip Options
#@markdown Optional settings for the unzip process.
#@markdown ---
#@markdown Select if you only want to list contents without extracting:
list_contents_only_param_str = "False" #@param ["False", "True"] # <<< Dropdown List
#@markdown ---
#@markdown **أ¢ع‘آ أ¯آ¸عˆ Warning:** Select "True" below to **delete the original zip file** after successful extraction. Use with caution!
delete_zip_after_param_str = "False"  #@param ["False", "True"]  # <<< Dropdown List
#@markdown ---

import zipfile
import os
import shutil

# --- Convert String Dropdown Selections to Boolean Values ---
list_contents_only = (list_contents_only_param_str == "True")
delete_zip_after = (delete_zip_after_param_str == "True")

# --- Debug Prints (Uncommented) ---
print(f"Debug Info: list_contents_only_param_str = '{list_contents_only_param_str}', converted to boolean: {list_contents_only}")
print(f"Debug Info: delete_zip_after_param_str = '{delete_zip_after_param_str}', converted to boolean: {delete_zip_after}")
# --- End Debug Prints ---


# --- FUNCTION DEFINITION (expects boolean arguments) ---
def unzip_colab_file(zip_path,
                     extract_to_path,
                     password=None,
                     list_contents_only=False, # Expects boolean
                     delete_zip_after=False): # Expects boolean
    """
    Unzips a file in Google Colab with added functionality.
    Includes option to delete zip after successful extraction.
    """
    print(f"--- Processing Zip File ---")
    print(f"Source: '{zip_path}'")
    print(f"Destination: '{extract_to_path}'")
    if password:
        print("Password: [Provided]")
    # Use the boolean args directly inside the function
    if list_contents_only:
        print("Mode: Listing contents only")
    if delete_zip_after:
        print("Option: Delete zip after success requested")

    # 1. Check if the zip file exists
    if not os.path.exists(zip_path):
        print(f"أ¢â€Œإ’ Error: Zip file not found at '{zip_path}'")
        return False
    if not os.path.isfile(zip_path):
         print(f"أ¢â€Œإ’ Error: Source path '{zip_path}' is not a file.")
         return False
    print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Passed initial file checks.")

    # 2. Ensure the extraction directory exists
    try:
        os.makedirs(extract_to_path, exist_ok=True)
        print(f"أ¢إ“â€‌أ¯آ¸عˆ Extraction directory ensured: '{extract_to_path}'")
    except OSError as e:
        print(f"أ¢â€Œإ’ Error: Could not create extraction directory '{extract_to_path}': {e}")
        return False
    print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Passed directory creation check.")

    operation_completed_successfully = False
    try:
        pwd_bytes = password.encode('utf-8') if password else None
        print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: About to open ZipFile: '{zip_path}'")
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Successfully opened ZipFile.")
            print(f"أ¢إ“â€‌أ¯آ¸عˆ Opened zip file successfully.")

            if list_contents_only: # Use boolean directly
                print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Entering list_contents_only block.")
                print("\n--- Contents ---")
                count = 0
                for file_name in zip_ref.namelist():
                    print(f"- {file_name}")
                    count += 1
                print("----------------")
                print(f"أ¢إ“â€‌أ¯آ¸عˆ Listing complete ({count} items).")
                operation_completed_successfully = True

            else: # Extracting
                print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Preparing to extract.")
                print(f"أ¢عˆآ³ Starting extraction...")
                zip_ref.extractall(path=extract_to_path, pwd=pwd_bytes)
                print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: extractall command finished.")
                print(f"أ¢إ“â€‌أ¯آ¸عˆ Extraction successful to '{extract_to_path}'.")
                operation_completed_successfully = True

    except zipfile.BadZipFile as e:
        print(f"أ¢â€Œإ’ DEBUG: Caught BadZipFile error.")
        print(f"أ¢â€Œإ’ Error: '{zip_path}' is not a valid zip file or is corrupted: {e}")
    except RuntimeError as e:
        print(f"أ¢â€Œإ’ DEBUG: Caught RuntimeError.")
        if 'password required' in str(e).lower() or 'bad password' in str(e).lower():
             print(f"أ¢â€Œإ’ Error: Password required or incorrect for '{zip_path}'. Please provide the correct password via the form.")
        else:
             print(f"أ¢â€Œإ’ Error during extraction: {e}")
    except FileNotFoundError as e:
         print(f"أ¢â€Œإ’ DEBUG: Caught FileNotFoundError during zip operation.")
         print(f"أ¢â€Œإ’ Error: Zip file disappeared during operation: '{zip_path}': {e}")
    except Exception as e:
        print(f"أ¢â€Œإ’ DEBUG: Caught generic Exception during zip operation.")
        print(f"أ¢â€Œإ’ An unexpected error occurred: {e}")
    finally:
        print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Reached finally block.")
        print(f"--- Zip processing finished ---")

    # Deletion logic uses the boolean delete_zip_after argument
    if operation_completed_successfully and delete_zip_after and not list_contents_only:
         print(f"--- Optional Deletion ---")
         try:
             print(f"أ¢عˆآ³ Deleting zip file: '{zip_path}'...")
             os.remove(zip_path)
             print(f"أ¢إ“â€‌أ¯آ¸عˆ Zip file deleted successfully.")
         except FileNotFoundError:
              print(f"أ¢ع‘آ أ¯آ¸عˆ Warning: Could not delete zip file '{zip_path}' after extraction: File was not found.")
         except OSError as e:
              print(f"أ¢ع‘آ أ¯آ¸عˆ Warning: Could not delete zip file '{zip_path}' after extraction: {e}")
    elif delete_zip_after and list_contents_only:
         print("   (Skipping deletion because 'List contents only' was selected)")
    elif delete_zip_after and not operation_completed_successfully:
         print("   (Skipping deletion because the primary unzip/list operation failed)")

    return operation_completed_successfully


# --- FUNCTION CALL (passing the CONVERTED boolean values) ---
print("\n--- Running Unzip with Form Parameters ---")

effective_password = password_param if password_param else None

# Call the function, passing the CONVERTED boolean values
success = unzip_colab_file(
    zip_path=zip_path_param,
    extract_to_path=extract_to_path_param,
    password=effective_password,
    list_contents_only=list_contents_only, # Use the boolean variable
    delete_zip_after=delete_zip_after     # Use the boolean variable
)

# --- Final status message ---
if success:
    print("\nأ¢إ“â€¦ Unzip/List operation reported success.")
else:
    print("\nأ¢â€Œإ’ Unzip/List operation reported failure.")
```

```python
# CELL 2 - FIXED CODE FOR BATCH UNRAR
#@markdown # ظ‹ع؛â€œآ¦ Unrar Utility for Google Colab (Batch Mode)
#@markdown ---
#@markdown ### ظ‹ع؛â€œظ¾ Input & Output Paths
#@markdown Enter the full path to the **folder containing your .rar files**.
input_folder_param = "/content/JDLDownloads" #@param {type:"string"}
#@markdown Enter the base directory where files should be extracted.
#@markdown Each RAR will be extracted into a subfolder named after the RAR file.
extract_to_base_path_param = "/content/extracted" #@param {type:"string"}
#@markdown Enter the password for the RAR files, if any. Leave blank if no password.
password_param = "" #@param {type:"string"}
#@markdown ---
#@markdown ### أ¢ع‘â„¢أ¯آ¸عˆ Unrar Options
#@markdown Optional settings for the unrar process for each file.
#@markdown ---
#@markdown Select if you only want to list contents without extracting:
list_contents_only_param_str = "False" #@param ["False", "True"]
#@markdown ---
#@markdown **أ¢ع‘آ أ¯آ¸عˆ Warning:** Select "True" below to **delete the original rar file** after successful extraction. Use with caution!
delete_rar_after_param_str = "False" #@param ["False", "True"]
#@markdown ---

import os
import sys
import shutil
import time
import subprocess
from pathlib import Path

# --- Convert String Dropdown Selections to Boolean Values ---
list_contents_only = (list_contents_only_param_str == "True")
delete_rar_after = (delete_rar_after_param_str == "True")

# --- Log Level Settings ---
# Set to True for detailed debugging, False for normal operation
DEBUG_MODE = False

def log_debug(message):
    """Print debug messages only if DEBUG_MODE is enabled"""
    if DEBUG_MODE:
        print(f"ظ‹ع؛â€‌ع† DEBUG: {message}")

def log_info(message):
    """Print informational messages"""
    print(f"أ¢â€‍آ¹أ¯آ¸عˆ {message}")

def log_success(message):
    """Print success messages"""
    print(f"أ¢إ“â€¦ {message}")

def log_warning(message):
    """Print warning messages"""
    print(f"أ¢ع‘آ أ¯آ¸عˆ WARNING: {message}")

def log_error(message):
    """Print error messages"""
    print(f"أ¢â€Œإ’ ERROR: {message}")

# --- DIRECT UNRAR COMMAND FUNCTION ---
def unrar_extract_command(rar_path, extract_path, password=None):
    """
    Use the system unrar command directly instead of the rarfile library
    which might have compatibility issues with some environments.
    """
    # Create command
    cmd = ["unrar", "x"]  # x for extract with full paths

    # Add password if provided
    if password:
        cmd.extend([f"-p{password}"])
    else:
        cmd.extend(["-p-"])  # No password

    # Add options
    cmd.extend([
        "-o+",       # Overwrite existing files
        "-idq",      # Disable percentage indicator but show filenames
        rar_path,    # Source RAR file
        extract_path # Destination folder
    ])

    # Add trailing slash to extraction path for unrar
    if not extract_path.endswith(os.sep):
        cmd[-1] = extract_path + os.sep

    # Execute command
    try:
        log_debug(f"Running command: {' '.join(cmd)}")
        process = subprocess.Popen(
            cmd,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            universal_newlines=True
        )

        # Capture output in real-time
        output_lines = []
        for line in process.stdout:
            line = line.strip()
            if line:
                output_lines.append(line)
                if not line.startswith("Creating ") and not line.startswith("Extracting "):
                    print(f"  {line}")

        # Wait for process to finish
        process.wait()

        # Check return code
        if process.returncode == 0:
            log_success(f"Extraction completed successfully")
            return True, output_lines
        elif process.returncode == 10:
            log_error(f"No files to extract (archive might be empty)")
            return False, output_lines
        elif process.returncode == 3:
            log_error(f"CRC error - The archive is corrupt or password is incorrect")
            return False, output_lines
        elif process.returncode == 11:
            log_error(f"Wrong password provided")
            return False, output_lines
        else:
            stderr_output = process.stderr.read()
            log_error(f"Extraction failed with code {process.returncode}: {stderr_output}")
            return False, output_lines

    except Exception as e:
        log_error(f"Failed to execute unrar command: {str(e)}")
        return False, []

# --- DIRECT UNRAR LIST FUNCTION ---
def unrar_list_command(rar_path, password=None):
    """List contents of a RAR file using direct unrar command"""
    cmd = ["unrar", "l"]  # l for list

    # Add password if provided
    if password:
        cmd.extend([f"-p{password}"])
    else:
        cmd.extend(["-p-"])  # No password

    cmd.append(rar_path)  # Source RAR file

    try:
        process = subprocess.run(
            cmd,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            universal_newlines=True,
            check=False  # Don't raise exception on non-zero return code
        )

        if process.returncode == 0:
            print("\n--- Contents of archive ---")
            # Parse the output to extract just the file list
            output_lines = process.stdout.split('\n')
            file_list = []
            capture = False

            for line in output_lines:
                if '-----------' in line:
                    if not capture:
                        capture = True  # Start capturing after the first dashed line
                        continue
                    else:
                        break  # Stop capturing after the second dashed line

                if capture and line.strip() and not 'Name' in line:
                    # Try to extract just the filename - unrar output has columns
                    parts = line.split()
                    if len(parts) >= 5:
                        # The filename is typically everything after the date/time and size
                        filename = ' '.join(parts[5:])
                        file_list.append(filename)
                        print(f"- {filename}")

            if not file_list:
                log_info("No files found in archive")
            else:
                log_success(f"Listed {len(file_list)} items in the archive")
            return True, file_list

        elif process.returncode == 10:
            log_error("No files to list (archive might be empty)")
            return False, []
        elif process.returncode == 11:
            log_error("Wrong password provided")
            return False, []
        else:
            log_error(f"Failed to list contents: {process.stderr}")
            return False, []

    except Exception as e:
        log_error(f"Failed to execute unrar list command: {str(e)}")
        return False, []

# --- MAIN UNRAR FUNCTION ---
def unrar_colab_file(rar_path,
                     extract_to_path,
                     password=None,
                     list_contents_only=False,
                     delete_rar_after=False):
    """
    Process a RAR file by directly calling the unrar command.
    This avoids issues with the rarfile Python library.
    """
    rar_filename = os.path.basename(rar_path)
    print(f"\n--- Processing RAR File: '{rar_filename}' ---")

    # Display operation parameters
    print(f"Source: '{rar_path}'")
    if not list_contents_only:
        print(f"Destination: '{extract_to_path}'")
    if password:
        print("Password: [Provided]")
    print(f"Mode: {'Listing contents only' if list_contents_only else 'Extraction'}")
    if delete_rar_after and not list_contents_only:
        print("Option: Delete RAR after successful extraction")

    # Validate source file exists
    if not os.path.exists(rar_path):
        log_error(f"RAR file not found at '{rar_path}'")
        return False
    if not os.path.isfile(rar_path):
        log_error(f"Source path '{rar_path}' is not a file")
        return False
    log_debug(f"Source file validation passed for '{rar_filename}'")

    # Create extraction directory if needed and not just listing
    if not list_contents_only:
        try:
            # Create directory with exist_ok to avoid race conditions
            os.makedirs(extract_to_path, exist_ok=True)

            # Verify directory was created and is writable
            if not os.path.isdir(extract_to_path):
                log_error(f"Failed to create extraction directory '{extract_to_path}'")
                return False

            # Quick write test to verify permissions
            test_file_path = os.path.join(extract_to_path, '.write_test')
            try:
                with open(test_file_path, 'w') as f:
                    f.write('test')
                os.remove(test_file_path)
            except (IOError, OSError) as e:
                log_error(f"Extraction directory '{extract_to_path}' is not writable: {e}")
                return False

            log_success(f"Extraction directory ready: '{extract_to_path}'")
        except OSError as e:
            log_error(f"Could not create or access extraction directory '{extract_to_path}': {e}")
            return False
        log_debug(f"Directory creation check passed for '{rar_filename}'")

    operation_completed_successfully = False

    # Handle based on operation type
    if list_contents_only:
        # List contents using direct unrar command
        operation_completed_successfully, _ = unrar_list_command(rar_path, password)
    else:
        # Extract using direct unrar command
        print(f"أ¢عˆآ³ Starting extraction...")
        operation_completed_successfully, _ = unrar_extract_command(rar_path, extract_to_path, password)

        # Verify extraction results
        if operation_completed_successfully:
            try:
                # Check if files were actually extracted
                extracted_items = os.listdir(extract_to_path)
                if not extracted_items:
                    log_warning(f"The directory '{extract_to_path}' is EMPTY after extraction. The archive might be empty or this could be part of a multi-part archive.")
                else:
                    num_to_show = min(10, len(extracted_items))
                    log_success(f"Extracted {len(extracted_items)} items")
                    log_debug(f"First {num_to_show} extracted items: {', '.join(extracted_items[:num_to_show])}")
            except Exception as e_list:
                log_warning(f"Could not verify extraction results: {e_list}")

    # Handle file deletion if requested
    if operation_completed_successfully and delete_rar_after and not list_contents_only:
        print(f"--- Optional Deletion for '{rar_filename}' ---")
        try:
            print(f"أ¢عˆآ³ Deleting RAR file...")
            os.remove(rar_path)
            log_success(f"RAR file deleted successfully")
        except FileNotFoundError:
            log_warning(f"Could not delete RAR file: File not found")
        except OSError as e:
            log_warning(f"Could not delete RAR file: {e}")
    elif delete_rar_after and list_contents_only:
        log_info(f"Skipping deletion because 'List contents only' was selected")
    elif delete_rar_after and not operation_completed_successfully:
        log_info(f"Skipping deletion because the operation failed or was incomplete")

    return operation_completed_successfully

# --- BATCH PROCESSING LOGIC ---
print("\n=== BATCH UNRAR UTILITY ===")
print(f"Mode: {'LISTING ONLY' if list_contents_only else 'EXTRACTION'}")

effective_password = password_param if password_param else None
files_processed = 0
files_successfully_processed = 0
files_failed = 0

if not os.path.exists(input_folder_param):
    log_error(f"Input folder not found at '{input_folder_param}'")
elif not os.path.isdir(input_folder_param):
    log_error(f"Input path '{input_folder_param}' is not a folder")
else:
    print(f"\nScanning folder: '{input_folder_param}' for .rar files...")
    found_rar_files_in_scan = False

    # Check if unrar command is available
    try:
        subprocess.run(["unrar", "--version"],
                      stdout=subprocess.DEVNULL,
                      stderr=subprocess.DEVNULL,
                      check=True)
    except (subprocess.SubprocessError, FileNotFoundError):
        log_error("The 'unrar' command is not available. Make sure you ran the setup cell first.")
        print("Trying to install unrar automatically...")
        try:
            subprocess.run(["apt-get", "update", "-qq"], check=True)
            subprocess.run(["apt-get", "install", "-y", "unrar"], check=True)
            log_success("Unrar installed successfully")
        except Exception as e:
            log_error(f"Failed to automatically install unrar: {e}")
            print("Please run the setup cell again before running this cell.")
            sys.exit(1)

    # Improved RAR file detection and sorting
    try:
        # Get all files in the directory
        all_files = os.listdir(input_folder_param)

        # Filter RAR files with improved detection (case-insensitive)
        rar_files = [f for f in all_files if f.lower().endswith('.rar') and
                    os.path.isfile(os.path.join(input_folder_param, f))]

        # Sort to process multi-part archives more intelligently
        # This handles both part01.rar and filename.part1.rar formats
        def sort_key(filename):
            # Convert to lowercase for case-insensitive sorting
            filename = filename.lower()
            # Check if this is a part file
            if '.part' in filename:
                # Extract the part number if it exists
                try:
                    # Handle formats like name.part01.rar, name.part1.rar
                    part_str = filename.split('.part')[1].split('.')[0]
                    # Remove leading zeros for numeric comparison
                    return (0, int(part_str.lstrip('0') or '0'))
                except (IndexError, ValueError):
                    # If we can't parse the part number, sort alphabetically
                    return (1, filename)
            # For non-part files, sort them before part files
            return (2, filename)

        rar_files.sort(key=sort_key)

        if not rar_files:
            print("No .rar files found in the specified folder.")
        else:
            found_rar_files_in_scan = True
            print(f"Found {len(rar_files)} RAR file(s) to process")

            # Process each RAR file
            for item_name in rar_files:
                current_rar_path = os.path.join(input_folder_param, item_name)
                files_processed += 1

                print(f"\n>>> Processing RAR file #{files_processed}/{len(rar_files)}: '{item_name}' <<<")

                # Create a clean subfolder name based on the archive name
                archive_name_without_ext = os.path.splitext(item_name)[0]
                # Remove .partX from multipart archives for a cleaner folder name
                if '.part' in archive_name_without_ext.lower():
                    base_parts = archive_name_without_ext.split('.part')
                    archive_name_without_ext = base_parts[0]

                current_extract_to_path = os.path.join(extract_to_base_path_param, archive_name_without_ext)

                success = unrar_colab_file(
                    rar_path=current_rar_path,
                    extract_to_path=current_extract_to_path,
                    password=effective_password,
                    list_contents_only=list_contents_only,
                    delete_rar_after=delete_rar_after
                )

                if success:
                    files_successfully_processed += 1
                else:
                    files_failed += 1
                print(f"--- Finished processing '{item_name}' ---")

    except Exception as e_listdir:
        log_error(f"Error scanning input folder '{input_folder_param}': {e_listdir}")

# --- FINAL SUMMARY ---
print("\n=== BATCH OPERATION SUMMARY ===")
if files_processed > 0:
    print(f"Total RAR files processed: {files_processed}")
    action_verb = "listed" if list_contents_only else "extracted"
    print(f"Successfully {action_verb}: {files_successfully_processed}")
    print(f"Failed operations: {files_failed}")
    if files_successfully_processed > 0 and not list_contents_only:
        print(f"Output base directory: '{extract_to_base_path_param}'")
        if files_failed > 0:
            print("أ¢ع‘آ أ¯آ¸عˆ Some operations failed. Check the logs above for details.")
    elif files_successfully_processed > 0 and list_contents_only:
        print("Listing operations complete.")
else:
    if os.path.exists(input_folder_param) and os.path.isdir(input_folder_param):
        print("No .rar files were found in the specified folder.")

print("\nأ¢إ“â€¦ Batch operation completed.")
```

```python
#@markdown # ظ‹ع؛â€œآ¦ Tar Utility for Google Colab
#@markdown ---
#@markdown ### ظ‹ع؛â€œظ¾ Input & Output Paths
#@markdown Enter the full path to the tar file and the desired extraction directory.
tar_path_param = "/content/Mirrored_Files/The.Summer.I.Turned.Pretty.S02.1080p.AMZN.WEB-DL.DDP5.1.H.265-PrimeFix.tar" #@param {type:"string"}
extract_to_path_param = "/content/Mirrored_Files" #@param {type:"string"}
#@markdown ---
#@markdown ### أ¢ع‘â„¢أ¯آ¸عˆ Tar Options
#@markdown Optional settings for the tar process.
#@markdown ---
#@markdown Select if you only want to list contents without extracting:
list_contents_only_param_str = "False" #@param ["False", "True"] # <<< Dropdown List
#@markdown ---
#@markdown **أ¢ع‘آ أ¯آ¸عˆ Warning:** Select "True" below to **delete the original tar file** after successful extraction. Use with caution!
delete_tar_after_param_str = "False"  #@param ["False", "True"]  # <<< Dropdown List
#@markdown ---

import tarfile
import os
import shutil

# --- Convert String Dropdown Selections to Boolean Values ---
list_contents_only = (list_contents_only_param_str == "True")
delete_tar_after = (delete_tar_after_param_str == "True")

# --- Debug Prints (Uncommented) ---
print(f"Debug Info: list_contents_only_param_str = '{list_contents_only_param_str}', converted to boolean: {list_contents_only}")
print(f"Debug Info: delete_tar_after_param_str = '{delete_tar_after_param_str}', converted to boolean: {delete_tar_after}")
# --- End Debug Prints ---


# --- FUNCTION DEFINITION (expects boolean arguments) ---
def untar_colab_file(tar_path,
                     extract_to_path,
                     list_contents_only=False, # Expects boolean
                     delete_tar_after=False): # Expects boolean
    """
    Untars a file in Google Colab with added functionality.
    Includes option to delete tar after successful extraction.
    """
    print(f"--- Processing Tar File ---")
    print(f"Source: '{tar_path}'")
    print(f"Destination: '{extract_to_path}'")
    # Use the boolean args directly inside the function
    if list_contents_only:
        print("Mode: Listing contents only")
    if delete_tar_after:
        print("Option: Delete tar after success requested")

    # 1. Check if the tar file exists
    if not os.path.exists(tar_path):
        print(f"أ¢â€Œإ’ Error: Tar file not found at '{tar_path}'")
        return False
    if not os.path.isfile(tar_path):
         print(f"أ¢â€Œإ’ Error: Source path '{tar_path}' is not a file.")
         return False
    print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Passed initial file checks.")

    # 2. Ensure the extraction directory exists
    try:
        os.makedirs(extract_to_path, exist_ok=True)
        print(f"أ¢إ“â€‌أ¯آ¸عˆ Extraction directory ensured: '{extract_to_path}'")
    except OSError as e:
        print(f"أ¢â€Œإ’ Error: Could not create extraction directory '{extract_to_path}': {e}")
        return False
    print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Passed directory creation check.")

    operation_completed_successfully = False
    try:
        print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: About to open TarFile: '{tar_path}'")
        with tarfile.open(tar_path, 'r') as tar_ref:
            print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Successfully opened TarFile.")
            print(f"أ¢إ“â€‌أ¯آ¸عˆ Opened tar file successfully.")

            if list_contents_only: # Use boolean directly
                print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Entering list_contents_only block.")
                print("\n--- Contents ---")
                count = 0
                for file_name in tar_ref.getnames():
                    print(f"- {file_name}")
                    count += 1
                print("----------------")
                print(f"أ¢إ“â€‌أ¯آ¸عˆ Listing complete ({count} items).")
                operation_completed_successfully = True

            else: # Extracting
                print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Preparing to extract.")
                print(f"أ¢عˆآ³ Starting extraction...")
                tar_ref.extractall(path=extract_to_path)
                print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: extractall command finished.")
                print(f"أ¢إ“â€‌أ¯آ¸عˆ Extraction successful to '{extract_to_path}'.")
                operation_completed_successfully = True

    except tarfile.TarError as e:
        print(f"أ¢â€Œإ’ DEBUG: Caught TarError.")
        print(f"أ¢â€Œإ’ Error: '{tar_path}' is not a valid tar file or is corrupted: {e}")
    except FileNotFoundError as e:
         print(f"أ¢â€Œإ’ DEBUG: Caught FileNotFoundError during tar operation.")
         print(f"أ¢â€Œإ’ Error: Tar file disappeared during operation: '{tar_path}': {e}")
    except Exception as e:
        print(f"أ¢â€Œإ’ DEBUG: Caught generic Exception during tar operation.")
        print(f"أ¢â€Œإ’ An unexpected error occurred: {e}")
    finally:
        print(f"أ¢إ“â€‌أ¯آ¸عˆ DEBUG: Reached finally block.")
        print(f"--- Tar processing finished ---")

    # Deletion logic uses the boolean delete_tar_after argument
    if operation_completed_successfully and delete_tar_after and not list_contents_only:
         print(f"--- Optional Deletion ---")
         try:
             print(f"أ¢عˆآ³ Deleting tar file: '{tar_path}'...")
             os.remove(tar_path)
             print(f"أ¢إ“â€‌أ¯آ¸عˆ Tar file deleted successfully.")
         except FileNotFoundError:
              print(f"أ¢ع‘آ أ¯آ¸عˆ Warning: Could not delete tar file '{tar_path}' after extraction: File was not found.")
         except OSError as e:
              print(f"أ¢ع‘آ أ¯آ¸عˆ Warning: Could not delete tar file '{tar_path}' after extraction: {e}")
    elif delete_tar_after and list_contents_only:
         print("   (Skipping deletion because 'List contents only' was selected)")
    elif delete_tar_after and not operation_completed_successfully:
         print("   (Skipping deletion because the primary untar/list operation failed)")

    return operation_completed_successfully


# --- FUNCTION CALL (passing the CONVERTED boolean values) ---
print("\n--- Running Untar with Form Parameters ---")

# Call the function, passing the CONVERTED boolean values
success = untar_colab_file(
    tar_path=tar_path_param,
    extract_to_path=extract_to_path_param,
    list_contents_only=list_contents_only, # Use the boolean variable
    delete_tar_after=delete_tar_after     # Use the boolean variable
)

# --- Final status message ---
if success:
    print("\nأ¢إ“â€¦ Untar/List operation reported success.")
else:
    print("\nأ¢â€Œإ’ Untar/List operation reported failure.")
```

```python
#@markdown <br><center><img src='https://raw.githubusercontent.com/dropcreations/Essential-Google-Colab-Notebook/0b62cd8a437d3439a1dbece9119b73bfa890ad1a/cell_logos/jDownloader-Logo.svg' height="60" alt="JDownloader-logo"/></center>
#@markdown <center><h3><b>Install JDownloader</b></h3></center><br>

import os, uuid, re, IPython
import ipywidgets as widgets

from glob import glob
from IPython.display import HTML, clear_output
from google.colab import output

# MODIFIED: Set default email and password
Email = widgets.Text(placeholder="*Required", description="Email:", value="edward80alexan.der.8.3.4.0@gmail.com")
Password = widgets.Text(placeholder="*Required", description="Password:", value="sadeq") # !!! SECURITY WARNING: Hardcoding passwords is risky !!!
Device = widgets.Text(placeholder="Optional", description="Name:")

# MODIFIED: Set default SavePath and add the new default to options
SavePath = widgets.Dropdown(value="/content/JDLDownloads", options=["/content", "/content/Downloads", "/content/JDLDownloads", "/content/drive/My Drive"], description="Save Path:")


class MakeButton(object):
    def __init__(self, title, callback, style):
        self._title = title
        self._callback = callback
        self._style = style

    def _repr_html_(self):
        callback_id = 'button-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        if self._style != "":
            style_html = "p-Widget jupyter-widgets jupyter-button widget-button mod-" + self._style
        else:
            style_html = "p-Widget jupyter-widgets jupyter-button widget-button"

        template = """<button class="{style_html}" id="{callback_id}">{title}</button>
            <script>
            document.querySelector("#{callback_id}").onclick = (e) => {{
                google.colab.kernel.invokeFunction('{callback_id}', [], {{}})
                e.preventDefault();
            }};
            </script>"""
        html = template.format(title=self._title, callback_id=callback_id, style_html=style_html)
        return html

def MakeLabel(description, button_style):
    return widgets.Button(description=description, disabled=True, button_style=button_style)

def RefreshPath():
    current_options = [
        "/content",
        "/content/Downloads",
        "/content/JDLDownloads" # ADDED: Ensure custom path is always an option
    ]
    if os.path.exists("/content/drive/"):
        current_options.append("/content/drive/My Drive")
        if os.path.exists("/content/drive/Shared drives/"):
            current_options.extend(glob("/content/drive/Shared drives/*"))

    # Preserve current value if it's valid, otherwise default or set to first
    current_val = SavePath.value
    SavePath.options = list(dict.fromkeys(current_options)) # Remove duplicates, preserve order
    if current_val in SavePath.options:
        SavePath.value = current_val
    elif "/content/JDLDownloads" in SavePath.options: # MODIFIED: Fallback to new default
        SavePath.value = "/content/JDLDownloads"
    elif current_options:
        SavePath.value = current_options[0]


def LoginForm():
    clear_output()

    # MODIFIED: Set default values when login form is displayed
    Email.value = "edward80alexan.der.8.3.4.0@gmail.com"
    Password.value = "sadeq" # !!! SECURITY WARNING: Hardcoding passwords is risky !!!
    Device.value = "" # Or set a default device name if you wish, e.g., "ColabJDownloader"

    RefreshPath()
    # MODIFIED: Ensure the default path is selected if available
    if "/content/JDLDownloads" in SavePath.options:
        SavePath.value = "/content/JDLDownloads"
    elif SavePath.options: # Fallback if custom default not available for some reason
        SavePath.value = SavePath.options[0]


    display(HTML("<h3 style=\"font-family:Trebuchet MS;color:#4f8bd6;\">If you don't have an account yet, please register <a href=\"https://my.jdownloader.org/login.html#register\" target=\"_blank\">here</a>.</h3>"), HTML("<br>"), Email, Password, Device, SavePath)

    if not os.path.exists("/content/drive/"):
        display(HTML("*If you want to save in Google Drive please run the cell below."))

    display(HTML("<br>"), MakeButton("Login", CheckLogin, "info"))

    if os.path.isfile("/root/.JDownloader/cfg/org.jdownloader.api.myjdownloader.MyJDownloaderSettings.json"):
        display(MakeButton("Cancel", Show, "danger"))

def RestartForm():
    clear_output()
    display(MakeLabel("Restart Confirm?", ""), MakeButton("Confirm", Restart, "danger"), MakeButton("Cancel", Show, "warning"))

def ExitForm():
    clear_output()
    display(MakeLabel("Exit Confirm?", ""), MakeButton("Confirm", Exit, "danger"), MakeButton("Cancel", Show, "warning"))

def CheckLogin():
    try:
        if not Email.value.strip():
            ERROR = "Email field is empty."
            THROW_ERROR
        if not "@" in Email.value and not "." in Email.value:
            ERROR = "Email is an incorrect format."
            THROW_ERROR
        if not Password.value.strip():
            ERROR = "Password field is empty."
            THROW_ERROR
        if not bool(re.match("^[a-zA-Z0-9]+$", Device.value)) and Device.value.strip():
            ERROR = "Only alphanumeric are allowed for the device name."
            THROW_ERROR
        Login()
    except NameError: # THROW_ERROR is not defined, so catch generic NameError or Exception
        print(ERROR)
    except Exception as e: # Catch any other potential error during checks
        print(f"An unexpected error occurred: {e}")


def Login():
    clear_output()
    # ADDED: Handle the new default path /content/JDLDownloads
    if SavePath.value == "/content/JDLDownloads":
        get_ipython().system_raw("mkdir -p -m 666 /content/JDLDownloads")
        get_ipython().system_raw("echo '{\"defaultdownloadfolder\" : \"/content/JDLDownloads\"}' > /root/.JDownloader/cfg/org.jdownloader.settings.GeneralSettings.json")
    elif SavePath.value == "/content":
        get_ipython().system_raw("echo '{\"defaultdownloadfolder\" : \"/content\"}' > /root/.JDownloader/cfg/org.jdownloader.settings.GeneralSettings.json")
    elif SavePath.value == "/content/Downloads":
        get_ipython().system_raw("mkdir -p -m 666 /content/Downloads")
        get_ipython().system_raw("echo '{\"defaultdownloadfolder\" : \"/content/Downloads\"}' > /root/.JDownloader/cfg/org.jdownloader.settings.GeneralSettings.json")
    else: # Handles GDrive paths or any other valid path
        get_ipython().system_raw("echo '{\"defaultdownloadfolder\" : \"" + SavePath.value + "\"}' > /root/.JDownloader/cfg/org.jdownloader.settings.GeneralSettings.json")

    if Device.value.strip() == "":
        Device.value = Email.value # Default device name to email if not provided
    get_ipython().system_raw("pkill -9 -e -f java")
    get_ipython().system_raw("echo '{\"email\" : \"'" + Email.value + "'\", \"password\" : \"'" + Password.value + "'\", \"devicename\" : \"'" + Device.value + "'\", \"directconnectmode\" : \"LAN\"}' > /root/.JDownloader/cfg/org.jdownloader.api.myjdownloader.MyJDownloaderSettings.json")
    get_ipython().system_raw("java -jar /root/.JDownloader/JDownloader.jar -norestart -noerr -r &")
    Show()

def Restart():
    get_ipython().system_raw("pkill -9 -e -f java")
    get_ipython().system_raw("java -jar /root/.JDownloader/JDownloader.jar -norestart -noerr -r &")
    Show()

def Exit():
    get_ipython().system_raw("pkill -9 -e -f java")
    clear_output()
    display(MakeButton("Start", Restart, "info"))

def Show():
    clear_output()
    display(MakeLabel("Control Panel", ""), HTML("<h3 style=\"font-family:Trebuchet MS;color:#4f8bd6;\">You can login to the WebUI by clicking <a href=\"https://my.jdownloader.org/\" target=\"_blank\">here</a>.</h3>"), HTML("<h4 style=\"font-family:Trebuchet MS;color:#4f8bd6;\">If the server didn't showup in 30 sec. please re-login.</h4>"), HTML("<br>"), MakeButton("Re-Login", LoginForm, "info"), MakeButton("Restart", RestartForm, "warning"), MakeButton("Exit", ExitForm, "danger"))


if not os.path.isfile("/root/.JDownloader/JDownloader.jar"):
    clear_output()
    display(MakeLabel("Installing in Progress", "warning"))
    get_ipython().system_raw("rm -rf /content/sample_data/ && apt install openjdk-8-jre-headless -qq -y && mkdir -p -m 666 /root/.JDownloader && wget -q http://installer.jdownloader.org/JDownloader.jar -O /root/.JDownloader/JDownloader.jar && java -jar /root/.JDownloader/JDownloader.jar -norestart -h")
    LoginForm()
elif not os.path.isfile("/root/.JDownloader/cfg/org.jdownloader.api.myjdownloader.MyJDownloaderSettings.json"):
    LoginForm()
else: Show()
```
